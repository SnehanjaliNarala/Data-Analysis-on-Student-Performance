## Data Analysis on Student Performance

```python
##Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##Data loading
df=pd.read_csv("Student.csv")
df
```

## Data Modification

```python
##Changing the column name from 'Passed' to 'Result'
df.columns=df.columns.str.replace('Passed','Result')
df
```

```python
##Changing the values in 'Result' Column
df.loc[df['Result']=='Yes','Result']='Pass'
df.loc[df['Result']=='No','Result']='Fail'
df
```

```python
##Changing the values in 'Student ID' Column
df['Student ID'] = [i for i in range(1, 40001)]
df
```

```python
## Replacing "NaN" values in 3 columns with 0
df.loc[df[['Study Hours per Week', 'Attendance Rate', 'Previous Grades']].isnull().any(axis=1), 
       ['Study Hours per Week', 'Attendance Rate', 'Previous Grades']] = 0
df
```

```python
## Rounding off the values of the 3 columns and converting them to integer data type
df['Previous Grades'] = df['Previous Grades'].astype(int)
df['Study Hours per Week'] = df['Study Hours per Week'].astype(int)
df['Attendance Rate'] = df['Attendance Rate'].astype(int)
df
```

```python
##saving cleaned and modified data
df.to_csv("student_modified.csv",index=False)
```

## Visualization of Student Results

```python
df_r.plot(kind='bar')
```

## Visualization of Student Attendance Rate

```python
df['Attendance Rate']
```

```python
bins = [0, 50, 65, 75, 100]  # Define the edges of the bins
labels = ['Below 50', '50 to 65', '65 to 75', 'Above 75']  # Define the labels for the bins

# Categorize the 'Attendance Rate' column into the specified bins
df['Attendance Category'] = pd.cut(df['Attendance Rate'], bins=bins, labels=labels, right=False)

# Group by the 'Attendance Category' and get the count of each category
grouped_df = df.groupby('Attendance Category').size().reset_index(name='Count')

# Display the grouped DataFrame
print(grouped_df)
```

```python
plt.bar(grouped_df['Attendance Category'],grouped_df['Count'])
plt.show()
```

## Visualization of Previous Grade among Students

```python
# Defining the bins and labels
bins = [0, 40, 50, 60, 75, 85, 90, 100,200]
labels = ['F', 'E', 'D', 'C', 'B', 'B+', 'A','A+']

# Categorizing the grades into bins
df['category1'] = pd.cut(df['Previous Grades'], bins=bins, labels=labels)

# Counting the number of students in each category
counts1 = df['category1'].value_counts()

print(counts1)
```

```python
plt.figure(figsize=(7,7))
counts1.plot(kind="pie",autopct='%1d%%')
plt.title('Distribution of Grades among students',fontsize=18)
plt.ylabel("Percentage of students")
plt.show()
```

## Visualization on Study Hours Per Week

```python
df['Study Hours per Week']
```

```python
bins = [0,10,20]  # Define the edges of the bins
labels = ['Below 10','Above 10']  # Define the labels for the bins

# Categorize the 'Attendance Rate' column into the specified bins
df['Studyhours Category'] = pd.cut(df['Study Hours per Week'], bins=bins, labels=labels, right=False)

# Group by the 'Attendance Category' and get the count of each category
df_2= df.groupby('Studyhours Category').size().reset_index(name='Count')

# Display the grouped DataFrame
print(df_2)
```

```python
plt.bar(df_2['Studyhours Category'],df_2['Count'])
plt.xlabel('Study hours')
plt.ylabel('Number of Students')
plt.title('Study Hours per Week')
plt.show()
```

## Visualization of Student Attendance Rates Distribution

```python
# Define the bins and labels
bins = [-float('inf'), 65, 75, float('inf')]
labels = ['Below 65', '65 to 75', 'Above 75']

# Categorize the attendance rates into bins
df['category'] = pd.cut(df['Attendance Rate'], bins=bins, labels=labels)

# Count the number of students in each category
counts = df['category'].value_counts()

print(counts)
```

```python
counts.plot(kind="pie",autopct='%1.1f%%',wedgeprops=dict(width=0.5))
plt.title('Distribution of Attendance Rates',fontsize=18)
plt.show()
```
